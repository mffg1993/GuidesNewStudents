## 1. ðŸ“ˆ Line Plot with Error Bars

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 20)
y = np.sin(x)
error = 0.1 + 0.1 * np.random.rand(len(x))

plt.errorbar(x, y, yerr=error, fmt='o-', capsize=4)
plt.title("Line Plot with Error Bars")
plt.xlabel("Position (mm)")
plt.ylabel("Signal (a.u.)")
plt.grid(True)
plt.show()
```

## 2. ðŸŸ¢ Scatter Plot with Colormap

```python
x = np.random.rand(100)
y = np.random.rand(100)
c = np.sqrt((x - 0.5)**2 + (y - 0.5)**2)

plt.scatter(x, y, c=c, cmap='viridis')
plt.colorbar(label="Distance from center")
plt.title("Scatter Plot with Color Encoding")
plt.xlabel("x")
plt.ylabel("y")
plt.axis('equal')
plt.show()
```

## 3. ðŸ“Š Bar Chart (Grouped and Stacked)

```python
labels = ['Laser A', 'Laser B', 'Laser C']
power_day = [3.1, 2.5, 4.0]
power_night = [2.8, 2.9, 3.5]

x = np.arange(len(labels))
width = 0.35

plt.bar(x - width/2, power_day, width, label='Day')
plt.bar(x + width/2, power_night, width, label='Night')
plt.xticks(x, labels)
plt.ylabel('Power (mW)')
plt.title('Laser Output Comparison')
plt.legend()
plt.grid(True, axis='y')
plt.show()
```

## 4. ðŸ“‰ Histogram

```python
data = np.random.normal(loc=0, scale=1, size=1000)

plt.hist(data, bins=30, color='skyblue', edgecolor='k')
plt.title("Histogram of Noise Distribution")
plt.xlabel("Voltage (V)")
plt.ylabel("Counts")
plt.grid(True)
plt.show()
```

## 6.  ðŸŒ Contour Plot

```python
X, Y = np.meshgrid(np.linspace(-3, 3, 100), np.linspace(-3, 3, 100))
Z = np.exp(-(X**2 + Y**2))

plt.contourf(X, Y, Z, levels=20, cmap='plasma')
plt.colorbar(label='Intensity')
plt.title("Contour Plot of Gaussian Beam")
plt.xlabel("x")
plt.ylabel("y")
plt.axis('equal')
plt.show()
```

## 7. ðŸŒ„ 3D Surface Plot

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
Z = np.exp(-(X**2 + Y**2))
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title("3D Surface of Beam Profile")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("Intensity")
plt.show()
```

## 8. ðŸ§­ Polar Plot

```python
theta = np.linspace(0, 2*np.pi, 100)
r = np.abs(np.sin(2*theta))

plt.polar(theta, r)
plt.title("Polar Plot (e.g. Radiation Pattern)")
plt.show()
```

## 9. ðŸ§­ Quiver (Vector Field) Plot

```python
X, Y = np.meshgrid(np.linspace(-1, 1, 10), np.linspace(-1, 1, 10))
U = -Y
V = X

plt.quiver(X, Y, U, V)
plt.title("Vector Field (e.g. Optical Flow)")
plt.xlabel("x")
plt.ylabel("y")
plt.axis('equal')
plt.grid(True)
plt.show()
```

## 10. ðŸ–¼ï¸ Displaying Images

```python
from scipy.ndimage import gaussian_filter
img = np.random.rand(100, 100)
blurred = gaussian_filter(img, sigma=2)

plt.imshow(blurred, cmap='gray')
plt.title("Simulated CCD Image")
plt.colorbar()
plt.axis('off')
plt.show()
```

## 12. ðŸŽ¨ Styling Tips and Exporting

```python
# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Styled plot
plt.figure(figsize=(6, 4))
plt.plot(x, y, label='Signal', linestyle='--', marker='s', color='darkred')

# Axis labels and title
plt.xlabel("Time (s)", fontsize=12)
plt.ylabel("Amplitude (a.u.)", fontsize=12)
plt.title("Styled Sine Wave Plot", fontsize=14)

# Grid, legend, and layout
plt.grid(True, which='both', linestyle=':', linewidth=0.5)
plt.legend()
plt.tight_layout()

# Export as high-resolution PNG
plt.savefig("styled_plot.png", dpi=300)
plt.show()
```
